```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
# Lista por comprensión donde guardo todos los valores en segundos, para operar
seconds_list = [(int(item['duration'].split(':')[0])*60 + int(item['duration'].split(':')[1])) for item in playlist]

# Saco longitud total y la formateo en mm:ss
print("Duración total: {}m {}s".format(int(sum(seconds_list)/60), int(sum(seconds_list)%60)))

# Opero sobre la lista de segundos y la uso para acceder a mi lista original
print("Canción más larga: \"{}\" ({})".format(playlist[seconds_list.index(max(seconds_list))]["title"],playlist[seconds_list.index(max(seconds_list))]["duration"]))
print("Canción más corta: \"{}\" ({})".format(playlist[seconds_list.index(min(seconds_list))]["title"],playlist[seconds_list.index(min(seconds_list))]["duration"]))
```
