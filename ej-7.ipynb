```python
# Amigo invisible
import random

friends = input("Ingrese los participantes (separados por coma):").split(", ") # Ingreso valores

def generatePairs(myList):
    if len(myList) %2 != 0: # Si son impares, soluciono o doy la chance de salir
        newValue = (input("Ingresaste una cantidad impar de personas! Ingresá una más o 0 para terminar"))
        if newValue == "0":
            return("La lista no se generó")
        myList.append(newValue)
    
    random.shuffle(myList) # Randomizo la lista
    results = []
    
    for i in range(0, int(len(myList))):
        gifter = myList[i] # Asigno la persona que va a regalar
        giftee = myList[(i + 1) % len(myList)] # Recibe el de al lado y si quien regala es último, el % len da 0 y da la vuelta.
        results.append((gifter, giftee)) # Agrego el resultado como tupla
    return results
    
print("Sorteo de amigo invisible:")
print(*(f"{item[0]} → {item[1]} \n" for item in generatePairs(friends)))
```
