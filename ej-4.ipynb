```python
email = input("Ingrese un email:")

def valid(text):
    arroba = text.find("@")
    return (
        text.count("@") == 1 and # Contiene exactamente un @.
        not text.startswith("@") and # Tiene al menos un carácter antes del @.
        "." in text[arroba:] and # Tiene al menos un punto (.) después del @.
        not text.endswith(".") and not text.startswith(".") and # No empieza ni termina con @ ni con ..
        len(text.split(".")[-1]) > 1 # La parte después del último punto tiene al menos 2 caracteres (el dominio).
    )

print("El email es válido" if valid(email) else "El email no es válido.")
```
